# How to parse a website with regex and urllib
 library modules, re and urllib, to parse paragraph data from a website. As
we saw, initially, when you use Python 3 and urllib to parse a website, you get all of the HTML data, like
using "view source" on a web page.

```python
import urllib.request
import re
```

```python
url = 'https://en.wikipedia.org/wiki/Wroc%C5%82aw'
req = urllib.request.Request(url)
resp = urllib.request.urlopen(req)
respData = resp.read()

paragraphs = re.findall(r'<p>(.*?)</p>',str(respData))

for eachP in paragraphs:
    print(eachP) #The output should be a bunch of paragraph data from our website.
```
